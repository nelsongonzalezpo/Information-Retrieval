We are going to define a function named "compare" whos mainly job will be to compare a certain path with a certain branch. This function will carry on tasks involve with context resemble when ever is invoked, we will proceed to explain the task carried out. First we are going to iterate through all the xmls file, and for each file we iterate through we will proceed to call the compare function. We are going to establish the path that we want the function to receive as an argument. 

Once we have establish the given path we will proceed to store this path in a variable named "path", we will make sure to change all the characters of the string to lowercase and split them with slash. Now we will proceed to compare the given nodes of the path with the nodes of the branch. Is important to mention that certain tasks will be carried out when we compare both node trees.

Fist of all we are going to check that the last node of the path is contain in the last node of the branch, if this is true then we will make sure that that all the nodes within the path until the last one are contain inside the branchs nodes until the last one too. If there are certain nodes in the branch that are not found in the path they will proceed to be removed. 

Depending on how similar the path and the branch are we will receive a percentage telling us wether the path matched the branch or not. This percentage mention earlier is what the context resemble subject is all about, the closer the number retured is to 1.0 it will mean that the branch and path are identical, if the number are close to 0 this means that the nodes contain within the path and the branch are not related at all.

```python
def compare(branch, path):
    path = path.lower().split("/")
    if path[len(path)-1] in branch[len(branch)-1]:
        if [x for x in branch[0:len(branch)-1] if x in path[0:len(path)-1]] == path[0:len(path)-1]:
            return (len(path)+1)/(len(branch)+1)
    return 0

for xml in xmls:
    if xml.file == "a_and_c.xml":
        for branch in xml.branches:
            num = compare(branch.b, "play/the")
            if num is not 0:
                print(num, branch.b, end='\n')
```
